**2. Определите матрицу признаков и переменную таргета. Разделите трейновый набор трейн и тест с помощью train_test_split. Зафиксируйте размер тестовой выборки 0.2, а рэндом стэйт равный 0.** 

```python
X = data.drop('target_class', axis = 1)
y = data['target_class']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

**3. Используя StandartScaler отмасштабируйте матрицу признаков трейн и тест.**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

**4. Обучите SVM Classifier на данных используя GridSearchCV: Попробуйте разные значения С, типа ядер, а так же значение гамма для rbf ядра. Подберите правильную метрику для оценки качества алгоритма. Убедитесь, что значения на тестровой и трейновой выборке сопоставимы и модель не переобучилась.**

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
model = SVC(kernel='rbf')
param_grid = {'C': [1, 5, 10, 50, 100, 500, 1000],
              'gamma': [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05, 0.1]}
grid = GridSearchCV(model, param_grid = param_grid)

grid.fit(X_train, y_train)
print(grid.best_params_)
```

```python
svc = SVC(kernel='rbf', C=50, gamma=0.05 )
```

```python
svc.fit(X_train, y_train)
```

```python
y_predicted_svc = svc.predict(X_train)
y_predicted_test_svc = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_predicted_svc)))
print('Testind-set accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_predicted_test_svc)))
```

```python
from sklearn.metrics import roc_auc_score

ROC_AUC = roc_auc_score(y_test, y_predicted_test_svc)

print('ROC AUC : {:.4f}'.format(ROC_AUC))
```

**5. Постройте confusion_matrix и classification report.**

```python
from sklearn.metrics import confusion_matrix
from sklearn import metrics
print (metrics.confusion_matrix(y_test, y_predicted_test_svc))
print (metrics.classification_report(y_test, y_predicted_test_svc))
```

**6. Обучите Decision tree classifier на тех же данных, используя GridSearchCV подберите гиперпараметры для модели. Используйте ту же метрику, что и в задаче с SVM. Убедитесь, что алгоритм не переобучился**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf_gini = DecisionTreeClassifier(criterion='gini', random_state=0)
param_grid = {'max_depth': [3, 5, 10, 12, 15, 20, 25]}

grid = GridSearchCV(clf_gini, param_grid = param_grid)

grid.fit(X_train, y_train)
print(grid.best_params_)
```

```python
clf_gini = DecisionTreeClassifier(criterion='gini', max_depth =5, random_state=0)
clf_gini.fit(X_train, y_train)
```

```python
y_predicted_gini = clf_gini.predict(X_train)
y_predicted_test_gini = clf_gini.predict(X_test)
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_predicted_gini)))
print('Testind-set accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_predicted_test_gini)))
```

```python
ROC_AUC = roc_auc_score(y_test, y_predicted_test_gini)

print('ROC AUC : {:.4f}'.format(ROC_AUC))
```

```python
clf_en = DecisionTreeClassifier(criterion='entropy', random_state=0)
param_grid = {'max_depth': [3, 5, 10, 12, 15, 20, 25]}

grid = GridSearchCV(clf_en, param_grid = param_grid)

grid.fit(X_train, y_train)
print(grid.best_params_)
```

```python
clf_en = DecisionTreeClassifier(criterion='entropy', max_depth =5, random_state=0)
clf_en.fit(X_train, y_train)
```

```python
y_predicted_en = clf_en.predict(X_train)
y_predicted_test_en = clf_en.predict(X_test)
```
