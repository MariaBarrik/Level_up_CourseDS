```python
X = df.drop(['Machine failure', "Air temperature [K]"], axis = 1)
y = df['Machine failure']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
for i in range(3,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    y_pred_train = knn.predict(X_train)
    print('Model accuracy score with k=',i,' : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
    print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train)))
    cm = confusion_matrix(y_test, y_pred)

    print('Confusion matrix\n\n', cm)

    print('\nTrue Positives(TP) = ', cm[0,0])

    print('\nTrue Negatives(TN) = ', cm[1,1])

    print('\nFalse Positives(FP) = ', cm[0,1])

    print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_predicted = model.predict(X_train)
y_predicted_test = model.predict(X_test)
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_predicted)))
print('Testind-set accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_predicted_test)))

```

```python
from sklearn import metrics
print (metrics.confusion_matrix(y_test, y_predicted_test))
print (metrics.classification_report(y_test, y_predicted_test))
```
